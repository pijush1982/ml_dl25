```python
def linear_regression_predict(xi, w, w0):
    """
    xi : list or 1D array of features for a single example
    w  : list or 1D array of weights
    w0 : bias term
    """
    n = len(xi)
    pred = w0

    for j in range(n):
        pred = w[j] * xi[j]

    return pred
```

## RMSE

```python
import numpy as np

def rmse(y, y_pred):
    error = y - y_pred
    se = error ** 2
    mse = se.mean()
    return np.sqrt(mse)
```
